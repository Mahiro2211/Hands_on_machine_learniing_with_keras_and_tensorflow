```python
from sklearn.impute import SimpleImputer
import pandas as pd
import matplotlib
```

```python
housing = pd.read_csv('housing.csv')
housing.isna().sum() , housing.describe()
```

```python

# 处理缺失值的三种方法
# housing = housing.dropna(subset=["total_bedrooms"]) # 放弃相应的区域
# housing = housing.drop("total_bedrooms",axis=1) # 放弃这个属性
# median = housing["total_bedrooms"].median() # 缺失的值设置为中位数，也可以是0，中位数平均数
# housing['total_bedrooms'].fillna(median,inplace=True)
# housing.fillna({'total_bedrooms':median},inplace=True) # recommended
# housing.isna().sum() , housing.describe()
```

```python
imputer = SimpleImputer(strategy='median')

housing_drop = housing.drop('ocean_proximity',axis=1)
imputer.fit(housing_drop)
```

```python
imputer.statistics_ , housing_drop.median().values, housing_drop.isna().sum()# 其实是计算了每个属性的中位数
```

```python
X = imputer.transform(housing_drop) # 可以把所有缺失值填充为中位数 返回的是一个numpy数组
housing_tr = pd.DataFrame(X,columns=housing_drop.columns,index=housing_drop.index) # 可以再转化为dataframe
```
